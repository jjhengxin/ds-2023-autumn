```python
import pandas as pd

file_path = 'literature_data.csv'  # 文件路径  
#data = pd.read_csv(file_path, encoding='GB2312', errors='ignore')
with open(file_path, 'r', encoding='gb2312', errors='ignore') as f:  
    content = pd.read_csv(f)
    # for row in content.iterrows():  
    #     print(row[1])
    timeI = dict()
    authorI = dict()

    # 处理出版年份列，并按照年份进行统计
    for i in range(content.shape[0]):
        if not content['publish_time'][i][4] == "年":
            continue
        print(content['publish_time'][i][:4])
        if content['publish_time'][i][:4] in timeI:
            timeI[content['publish_time'][i][:4]] += 1
        else :
            timeI[content['publish_time'][i][:4]] = 1
        
        if content['author'][i] in authorI:
            authorI[content['author'][i]] += 1
        else :
            authorI[content['author'][i]] = 1
    print(timeI.keys(),timeI.values())
    
```

```python
from openpyxl import Workbook

def write_dict_to_excel(data_dict, filename):
    # 创建一个新的Excel工作簿
    workbook = Workbook()

    # 获取活动的工作表
    worksheet = workbook.active

    # 将键值对逐行写入excel表格
    for key in data_dict.keys():
        myli = []
        myli.append(key)
        myli.append(data_dict[key])
        worksheet.append(myli)
    # 保存Excel文件
    workbook.save(filename)


# 调用函数将字典写入Excel文件
write_dict_to_excel(timeI, '出版年份.xlsx')
```
